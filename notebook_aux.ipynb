Scripts para juntar train y val y dividirlo en porci√≥n 80 20.

```python
import os
import shutil

TRAIN_DIR = "data/train"
VAL_DIR = "data/val"

for cluster in sorted(os.listdir(TRAIN_DIR)):
    train_cluster_dir = os.path.join(TRAIN_DIR, cluster)
    val_cluster_dir = os.path.join(VAL_DIR, cluster)
    for filename in os.listdir(val_cluster_dir):
        shutil.move(os.path.join(val_cluster_dir, filename), train_cluster_dir)
```

```python
import random

for cluster in sorted(os.listdir(TRAIN_DIR)):
    cluster_dir = os.path.join(TRAIN_DIR, cluster)
    num_images = len(os.listdir(cluster_dir))
    num_val_images = num_images // 5
    print(f"num_val_images {num_val_images}, total images {num_images} in cluster {cluster} ")
    for i in range(num_val_images):
        random_index = random.randint(0, num_images - 1)
        filename = os.listdir(cluster_dir)[random_index]
        shutil.move(os.path.join(cluster_dir, filename), os.path.join(VAL_DIR, cluster, filename))
        num_images -= 1
```
